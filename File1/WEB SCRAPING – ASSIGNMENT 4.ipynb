```python
No_of_attribute=[]
start=0
end=10
for page in range(start,end):
    NOA=driver.find_elements(By.XPATH,'//div[4][@class="col-span-3 flex items-center gap-2"]/span')
    for i in NOA:
        if i.text is None :
            No_of_attribute.append("--")
        else:
            No_of_attribute.append(i.text)
print(len(No_of_attribute),No_of_attribute)
```
